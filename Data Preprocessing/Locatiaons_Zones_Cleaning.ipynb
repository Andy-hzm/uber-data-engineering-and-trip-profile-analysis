**We consider that we should use borough to cover number of locations in it, so we need to use range of latitutes and longtitutes.**

```python
borough_ranges = {
    "Bronx": {"lat_min": 40.80, "lat_max": 40.90, "lon_min": -73.90, "lon_max": -73.85},
    "Brooklyn": {"lat_min": 40.63, "lat_max": 40.70, "lon_min": -73.95, "lon_max": -73.90},
    "Manhattan": {"lat_min": 40.75, "lat_max": 40.85, "lon_min": -74.00, "lon_max": -73.93},
    "Queens": {"lat_min": 40.68, "lat_max": 40.73, "lon_min": -73.85, "lon_max": -73.75},
    "Staten Island": {"lat_min": 40.57, "lat_max": 40.60, "lon_min": -74.20, "lon_max": -74.10},
}

borough_ranges_df = pd.DataFrame.from_dict(borough_ranges, orient='index')

borough_ranges_df.reset_index(inplace=True)
borough_ranges_df.rename(columns={'index': 'borough'}, inplace=True)
```

```python
borough_ranges_df["borough_id"] = range(1, len(borough_ranges_df) + 1)
```

```python
borough_ranges_df
```

```python
boroughs_file = "boroughs_with_geom.csv"
borough_ranges_df.to_csv(boroughs_file, index=False)
```

```python
zones = pd.read_csv('taxi_zones.csv')
```

```python
zones = zones.merge(borough_ranges_df, on="borough", how="left")
```

```python
zones = zones.dropna(subset=["borough_id"])

zones["borough_id"] = zones["borough_id"].astype(int)
```
