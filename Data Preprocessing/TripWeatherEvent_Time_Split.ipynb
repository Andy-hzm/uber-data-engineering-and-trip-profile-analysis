```python
# Split Column datetime to two columns: date and time in Trip data
import pandas as pd
from pytz import timezone

input_file = r"C:\Users\Dazhou Wu\Desktop\final_trip_data.csv"
output_file = "C:/Users/Dazhou Wu/Desktop/final_edit_trip_data.csv"

df = pd.read_csv(input_file)

new_york = timezone("America/New_York")

for col in ['request_datetime', 'on_scene_datetime', 'pickup_datetime', 'dropoff_datetime']:
    df[col] = pd.to_datetime(df[col].str.replace(" UTC", ""), utc=True).dt.tz_convert(new_york)

    df[f"{col}_date"] = df[col].dt.strftime('%Y-%m-%d') 
    df[f"{col}_time"] = df[col].dt.strftime('%H:%M')    

df.drop(columns=['request_datetime', 'on_scene_datetime', 'pickup_datetime', 'dropoff_datetime'], inplace=True)

df.rename(columns={
    'request_datetime_date': 'request_date',
    'request_datetime_time': 'request_time',
    'on_scene_datetime_date': 'on_scene_date',
    'on_scene_datetime_time': 'on_scene_time',
    'pickup_datetime_date': 'pickup_date',
    'pickup_datetime_time': 'pickup_time',
    'dropoff_datetime_date': 'dropoff_date',
    'dropoff_datetime_time': 'dropoff_time'
}, inplace=True)

df.to_csv(output_file, index=False)

print(f"Data Saved as: {output_file}")
```

```python
# Split Column datetime to two columns: date and time in Weather Daily data
file_path = r"C:\Users\Dazhou Wu\Desktop\Mine\master\Fall 24\Data Engineering\Final Project\dataset\weather_daily.csv"
weather_df = pd.read_csv(file_path)
weather_df.head()
```

```python
weather_df[['sunrise_date', 'sunrise_time']] = weather_df['sunrise (iso8601)'].str.split('T', expand=True)
weather_df[['sunset_date', 'sunset_time']] = weather_df['sunset (iso8601)'].str.split('T', expand=True)

# Drop original columns if no longer needed
weather_df.drop(columns=['sunrise (iso8601)', 'sunset (iso8601)','time'], inplace=True)
weather_df
```

```python
output_file = "C:/Users/Dazhou Wu/Desktop/final_weather_daily.csv"
weather_df.to_csv(output_file, index=False)

print(f"Data Saved as: {output_file}")
```

```python
# Split Column datetime to two columns: date and time in Weather Hourly
file_path = r"C:\Users\Dazhou Wu\Desktop\Mine\master\Fall 24\Data Engineering\Final Project\dataset\weather_hourly.csv"
weather_hourly_df = pd.read_csv(file_path)
weather_hourly_df.head()
```

```python
# Split Column datetime to two columns: date and time in Event
file_path = r"C:\Users\Dazhou Wu\Desktop\final_event.csv"
event_df = pd.read_csv(file_path)
event_df.head()
```

```python
event_df[['Start Date', 'Start Time']] = event_df['Start Date Time'].str.split(' ', expand=True)

# Split 'End Date Time' into 'End Date' and 'End Time'
event_df[['End Date', 'End Time']] = event_df['End Date Time'].str.split(' ', expand=True)

# Convert 'Start Date' and 'End Date' to the desired format (YYYY-MM-DD)
event_df['Start Date'] = pd.to_datetime(event_df['Start Date'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')
event_df['End Date'] = pd.to_datetime(event_df['End Date'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')

# Drop the original 'Start Date Time' and 'End Date Time' columns if no longer needed
event_df.drop(columns=['Start Date Time', 'End Date Time'], inplace=True)
event_df
```

```python
output_file = "C:/Users/Dazhou Wu/Desktop/final_event_edit.csv"
event_df.to_csv(output_file, index=False)

print(f"Data Saved as: {output_file}")
```
